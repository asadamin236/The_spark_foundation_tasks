## Drop the two columns id and species.

```python
X=Df.drop(['Id','Species'], axis=1)
```

## Apply the kmeans clustering on the dataset

```python
plt.rcParams['figure.figsize']=(10, 5)
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)
    
#plot the results
plt.plot(range(1, 11), wcss)
plt.title('The ELbo Chart', fontsize=15)
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Fitting our dataset with Kmeans clustring model with 3 number of clusters because our greater point on 3.

kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
Y = kmeans.fit_predict(X)
X=X.values

# Predicting clusters of the dataset and visualize on scatterlot. 

plt.figure(figsize=(15,8))
plt.scatter(X[Y==0, 0], X[Y==0, 1], s =20, c ='red', label='Iris-setosa')
plt.scatter(X[Y==1, 0], X[Y==1, 1], s =20, c ='green', label='Iris-versicolor')
plt.scatter(X[Y==2, 0], X[Y==2, 1], s =20, c ='blue', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
           s=100, c='yellow', label="Centroids")
plt.legend()
plt.show()
```
